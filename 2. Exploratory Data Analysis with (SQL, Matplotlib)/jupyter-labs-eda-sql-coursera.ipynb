## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1

##### Display the names of the unique launch sites  in the space mission


```python
def get_unique_launch_sites():
    sql = "select distinct LAUNCH_SITE from SPACEX"
    dataset = execute_sql_ret_df(sql)
    return dataset
get_unique_launch_sites().head()
```

### Task 2

##### Display 5 records where launch sites begin with the string 'CCA'


```python
def get_unique_launch_sites_starting_CCA():
    sql = "SELECT distinct LAUNCH_SITE from SPACEX where LAUNCH_SITE LIKE 'CCA%'"
    dataset = execute_sql_ret_df(sql)
    return dataset

get_unique_launch_sites_starting_CCA().head()
```

### Task 3

##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
def total_payload_mass_nasa():
    sql = "select sum(PAYLOAD_MASS__KG_) as total_payload_mass_kg from spacex where CUSTOMER like '%CRS%' "
    dataset = execute_sql_ret_df(sql)
    return dataset

total_payload_mass_nasa().head()
```

### Task 4

##### Display average payload mass carried by booster version F9 v1.1


```python
def average_payload_mass_F911():
    sql = "select avg(PAYLOAD_MASS__KG_) as avg_payload_mass_kg from spacex where BOOSTER_VERSION like '%F9 v1.1%'"
    dataset = execute_sql_ret_df(sql)
    return dataset

average_payload_mass_F911().head()
```

### Task 5

##### List the date when the first successful landing outcome in ground pad was acheived.

*Hint:Use min function*


```python
def first_successful_landing():
    sql = "SELECT min(DATE) as FIRST_SUCCESSFUL from SPACEX where MISSION_OUTCOME = 'Success'"
    dataset = execute_sql_ret_df(sql)
    return dataset

first_successful_landing().head()
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
def boosters_4k_to_6k():
    sql = "SELECT distinct BOOSTER_VERSION from SPACEX where LANDING__OUTCOME = 'Success (drone ship)' and PAYLOAD_MASS__KG_ >4000 and PAYLOAD_MASS__KG_ <6000"
    dataset = execute_sql_ret_df(sql)
    return dataset

boosters_4k_to_6k().head()
```

### Task 7

##### List the total number of successful and failure mission outcomes


```python
def successfull_failed_total():
    sql = "SELECT MISSION_OUTCOME, COUNT(MISSION_OUTCOME) as TOTAL from SPACEX group by MISSION_OUTCOME"
    dataset = execute_sql_ret_df(sql)
    return dataset

successfull_failed_total().head()
```

### Task 8

##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
def boosters_max_payload():
    sql = "SELECT DISTINCT BOOSTER_VERSION FROM SPACEX WHERE PAYLOAD_MASS__KG_ = (select max(PAYLOAD_MASS__KG_) from spacex)"
    dataset = execute_sql_ret_df(sql)
    return dataset

boosters_max_payload().head()
```

### Task 9

##### List the failed landing_outcomes in drone ship, their booster versions, and launch site names for in year 2015


```python
def landings_drone_2015():
    sql = "SELECT BOOSTER_VERSION,LAUNCH_SITE FROM SPACEX WHERE LANDING__OUTCOME = 'Failure (drone ship)' AND DATE_PART('YEAR', DATE) = 2015"
    dataset = execute_sql_ret_df(sql)
    return dataset

landings_drone_2015().head()
```

### Task 10

##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order


```python
def landing_rank_2k10_2k17():
    sql = "select LANDING__OUTCOME, COUNT(LANDING__OUTCOME) AS TOTAL from spacex where DATE > '2010-06-04' and DATE < '2017-03-20' GROUP BY LANDING__OUTCOME ORDER BY 2 DESC"
    dataset = execute_sql_ret_df(sql)
    return dataset

landing_rank_2k10_2k17().head(20)
```
